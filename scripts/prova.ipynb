```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_s_curve():
    M = 10
    N = 3000

    regret = np.ones(M)

    for s in range(1,M):
        r = 1
        for j in range(N):
            d = j/N
            miss = -d*s-0.1
            r = min(r, max(d/2+1/2, d/2+miss+1))
        regret[s] = r

    return regret

def lower_bound(s):
    return (1+s)/(1+2*s)

def gauss(s):
    return (s+2.5)/(2*s+3)
```

```python
c = plot_s_curve()

M = 10
s = np.linspace(0,M,M)

plt.plot(s,c)
plt.plot(s,lower_bound(s))
plt.plot(s,gauss(s))
```
