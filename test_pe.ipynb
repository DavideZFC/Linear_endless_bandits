```python
import numpy as np
from classes.environment import Environment
from classes.PE import PE

import matplotlib.pyplot as plt
```

```python
d = 5
n_arms = 100
T = 10000
m = 100
fhdj = 0

theta = np.array([0.9, -1., 1., -1., 1.1])

env = Environment(theta, sd=0.1, epsilon=0.1)

arms = np.random.normal(size=(n_arms,d))
```

```python
reward_vector = np.zeros(T)
policy = PE(arms_matrix=arms, T=T)


for t in range(T):
    arm = policy.pull_arm()
    arm_vector = arms[arm,:]
    reward = env.pull_arm(arm_vector)
    reward_vector[t] = reward
    policy.update(arm_vector, reward)
```

```python
plt.plot(reward_vector)
```
